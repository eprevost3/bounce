```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def y(t) :
    return(1 / 2 * 9.81 * t ** 2 + -5 * t)


t = np.arange(0, 1, .1)
pos = y(t)

plt.plot(t, pos)
plt.ylim([0, -1.2])
plt.show()
```

```python
def computePos(y, dy, deltaTime, yMin = -1e6, yMax = 1e6) :
    g = 9.81

    newDy = - dy + g * deltaTime
    newY = - y - dy * deltaTime + 1/2 * g * deltaTime ** 2

    if newY <= yMin : newY = yMin
    elif newY >= yMax : newY = yMax

    return([newY, newDy])

position = [0]
vitesse = [0]

y, dy = computePos(0, 0, 1e-3)
for k in range(1000) : 
    y, dy = computePos(y, dy, 1e-3)
    position.append(y), vitesse.append(dy)
```

```python
plt.figure(figsize = (15,10))
plt.plot(position)
plt.show()
```
